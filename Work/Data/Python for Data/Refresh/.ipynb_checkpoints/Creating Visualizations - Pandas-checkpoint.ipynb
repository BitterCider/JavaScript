# Pandas Visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\user\Downloads\Ice Cream Ratings.csv")
df = df.set_index('Date')
```

```python
df
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating',
        xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df['Flavor Rating'].plot(kind='barh', stacked=True)
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=400, c='y')
```

```python
df.plot.hist(bins=20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating',figsize = (10,10))
```
