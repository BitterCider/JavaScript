# Data Cleaning 

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\user\Downloads\Customer Call List.xlsx")
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df = df.drop(columns='Not_Useful_Column')
```

```python
#import re
#pattern = re.compile('[^A-Za-z]')
#df['Last_Name'] = df['Last_Name'].astype(str)
#df['Last_Name'] = df['Last_Name'].apply(lambda x: re.sub(pattern, '', x) if x is not None else x)

df['Last_Name'] = df['Last_Name'].str.strip('123./_')

df
```

```python
import re
#pattern = re.compile('[^0-9]')
#df['Phone_Number'] = df['Phone_Number'].astype(str)
#df['Phone_Number'] = df['Phone_Number'].apply(lambda x: re.sub(pattern, '', x) if x is not None else x)
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]', '', regex=True)
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str)

df['Phone_Number'] = df['Phone_Number'].apply(
    lambda x: f'{x[0:3]}-{x[3:6]}-{x[6:10]}' if '-' not in x else x)

pattern = re.compile('^[A-Za-z]+')

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: re.sub(pattern, '', x))
```

```python
df[['Street_Name', 'State', 'Zip_Code']] = df['Address'].str.split(',',n=2, expand=True)
df
```

```python
df[['Paying Customer', 'Do_Not_Contact']]=df[['Paying Customer', 'Do_Not_Contact']].replace('N', 'No')
df[['Paying Customer', 'Do_Not_Contact']]=df[['Paying Customer', 'Do_Not_Contact']].replace('Y', 'Yes')
```

```python
df = df.replace('N/a', '')
df = df.fillna('')
```

```python
df
```

```python
df.index
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Yes' or df.loc[x, 'Phone_Number'] == '--':
        df.drop(x, inplace=True)

df
```

```python
df.set_index('CustomerID', inplace=True)
```
